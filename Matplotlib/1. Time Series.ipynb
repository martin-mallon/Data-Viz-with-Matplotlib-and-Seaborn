```python
# libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#
from ipynb.fs.full.Style_sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)

# import data
def csv_to_data(file_path, parse_dates, index_col):
    df = pd.read_csv(file_path, parse_dates=parse_dates, index_col=index_col)
    return df

# data
cc = csv_to_data('Data\climate_change.csv', True, 'date')
cc.head()
```

```python
# line plot
def viz_line(ax, data, y, x, title):
    sns.lineplot(ax=ax, data=data, y=y, x=x, color=palette_blues[4])
    ax.set_title(title)
    return ax

# fig
fig, [ax_1, ax_2] = plt.subplots(nrows=2, ncols=1, figsize=(18,12), constrained_layout=True)

# data
cc_all = cc
cc_70 = cc['1970-01-01':'1970-12-31']

# viz
viz_line(ax_1, cc_all, 'co2', cc_all.index, 'CO2 levels')
viz_line(ax_2, cc_70, 'co2', cc_70.index, '1970s')

# show
plt.show()
```
