<h1>Scatterplot</h1>

```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

#
from ipynb.fs.full.Style_sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)

# import data
def csv_to_data(file_path, parse_dates, index_col):
    df = pd.read_csv(file_path, parse_dates=parse_dates, index_col=index_col)
    return df

# data
cc = csv_to_data('Data\climate_change.csv', True, 'date')
cc.head()
```

```python
# scatterplot
fig, ax_1 = plt.subplots(nrows=1, ncols=1, figsize=(15,10))
sns.scatterplot(ax=ax_1, data=cc, x='co2', y='relative_temp', hue=cc.index, legend=False)
plt.show()
```

```python
# viz_scatter
def viz_scatter(ax, data, x, y, hue, title, xlabel, ylabel):
    sns.scatterplot(ax=ax, data=data, x=x, y=y, hue=hue, legend=False)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    return ax

# viz
fig, ax_1 = plt.subplots(nrows=1, ncols=1, figsize=(15,10))
viz_scatter(ax_1, cc, 'co2', 'relative_temp', cc.index, 'CO2 vs Temperature', 'CO2', 'Temperature')
plt.show()
```
