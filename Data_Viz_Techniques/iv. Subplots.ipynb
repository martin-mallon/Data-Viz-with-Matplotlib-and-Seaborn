<h1>Subplots with <code>plt.subplots()</code></h1>

```python
# libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# style
from ipynb.fs.full.Style_sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)

# import data
def csv_to_data(file_path, parse_dates, index_col):
    df = pd.read_csv(file_path, parse_dates=parse_dates, index_col=index_col)
    return df

# line plot
def viz_line(ax, data, y, color, title, y_label, x_label):
    sns.lineplot(ax=ax, data=data, y=y, x=data.index, color=color)
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    return ax

# data
cc = csv_to_data('Data\climate_change.csv', True, 'date')

# fig subplots
fig, [ax_1, ax_2] = plt.subplots(nrows=2, ncols=1, figsize=(12,12), constrained_layout=True)
viz_line(ax_1, cc, 'relative_temp', palette_blues[4], 'CO2 levels', 'CO2', 'Date')
viz_line(ax_2, cc, 'co2', palette_blues[4], '1970s', 'CO2', 'Date')

# show
plt.show()
```
