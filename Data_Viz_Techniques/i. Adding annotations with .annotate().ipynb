<h1>Adding annotations with <code>.annotations()</code></h1>

```python
# libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# style
from ipynb.fs.full.Style_sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)

# import data
def csv_to_data(file_path, parse_dates, index_col):
    df = pd.read_csv(file_path, parse_dates=parse_dates, index_col=index_col)
    return df

# line plot
def viz_line(ax, data, y, color, title, y_label, x_label):
    sns.lineplot(ax=ax, data=data, y=y, x=data.index, color=color)
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    return ax

# data
cc = csv_to_data('Data\climate_change.csv', True, 'date')

# fig
fig, ax_1 = plt.subplots(nrows=1, ncols=1, figsize=(12,6), constrained_layout=True)

# viz
viz_line(ax_1, cc, 'relative_temp', palette_paired[1], 'Climate Change', 'Temp. C', 'Date')

# annotate
ax_1.annotate('Temp>1', 
              xy=(pd.Timestamp('2015-10-06'), 1),
              xytext=(pd.Timestamp('2008-10-06'), -0.2),
              arrowprops={'arrowstyle':'->', 'color':'black'}
             )

# show
plt.show()
```
