<h1>Saving Viz as <code>.png</code> and <code>.jpg</code></h1>

```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

#
from ipynb.fs.full.Style_sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)

# import data
def csv_to_data(file_path):
    df = pd.read_csv(file_path)
    return df

# data
mw = csv_to_data('Data\medalist_weights.csv')
cols = ['ID', 'Sport', 'Height']
rh = mw[cols].query('Sport == "Rowing" or Sport == "Gymnastics"')

# viz_hist
def viz_hist(ax, data, x, kde, title):
    
    # sort bins
    x_min = round(min(data[x]),0)
    x_max = round(max(data[x]),0)
    bins_n = int((x_max - x_min) / 5)
    
    # plot
    sns.histplot(ax=ax, data=data, x=x, kde=kde,
                 hue='Sport', stat='probability',
                 common_norm=False, element='step', bins=bins_n)
    ax.set_title(title)
    #ax.set_ylim(0, 1)
    
    # return
    return ax

# viz
fig, ax_1 = plt.subplots(nrows=1, ncols=1, figsize=(12,8), constrained_layout=True)
viz_hist(ax_1, rh, 'Height', True, 'Athlete Height Distributions')
plt.show()
```

<br>

<ul>
    <li><b><code>image.png / image.jpg</code></b> - Dictates quality and memory of file.</li>
    <br>
    <li><b><code>image.svg</code></b> - Produces vector graphic file where elements can be edited individually.</li>
    <br>
    <li><b><code>quality</code></b> - 0-100 value that further dictates quality of specified file.</li>
    <br>
    <li><b><code>dpi</code></b> - DOts per inch, dictates resolution.</li>
</ul>

```python
# image formats
fig.savefig('Viz_PNG.png')
fig.savefig('Viz_JPG.jpg')
fig.savefig('Viz_SVG.svg')
```
