<h1>Line plot</h1>

```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

#
from ipynb.fs.full.Style_sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)

# import data
def csv_to_data(file_path, parse_dates, index_col):
    df = pd.read_csv(file_path, parse_dates=parse_dates, index_col=index_col)
    return df

# data
cc = csv_to_data('Data\climate_change.csv', True, 'date')
cc.head()
```

```python
# line plot
fig, ax_1 = plt.subplots(nrows=1, ncols=1, figsize=(12,6), constrained_layout=True)
sns.lineplot(ax=ax_1, data=cc, x='date', y='relative_temp', color=palette_blues[4])
plt.show()
```

```python
# viz_line
def viz_line(ax, data, x, y, color, title, y_label, x_label):
    sns.lineplot(ax=ax, data=data, y=y, x=x, color=color)
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    return ax

# viz
fig, ax_1 = plt.subplots(nrows=1, ncols=1, figsize=(12,6), constrained_layout=True)
viz_line(ax_1, cc, 'date', 'relative_temp', palette_blues[4], 'Rising Temperatures', 'Temp. C', 'Date')
plt.show()
```
