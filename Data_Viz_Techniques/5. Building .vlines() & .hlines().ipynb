<h1>Building <code>.vlines()</code> & <code>.hlines()</code></h1>

```python
# libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# style
from ipynb.fs.full.Style_sheet import *
plt.style.use(style_use)
plt.rcParams.update(style_params)

# import data
def csv_to_data(file_path, parse_dates, index_col):
    df = pd.read_csv(file_path, parse_dates=parse_dates, index_col=index_col)
    return df

# line plot
def viz_line(ax, data, y, color, title, y_label, x_label):
    sns.lineplot(ax=ax, data=data, y=y, x=data.index, color=color)
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_ylim([ax_1.get_ylim()[0], ax_1.get_ylim()[1]])
    ax.set_xlim([ax_1.get_xlim()[0], ax_1.get_xlim()[1]])
    return ax

# data
cc = csv_to_data('Data\climate_change.csv', True, 'date')
cc.head()
```

```python
# fig
fig, ax_1 = plt.subplots(nrows=1, ncols=1, figsize=(12,6), constrained_layout=True)

# viz
viz_line(ax_1, cc, 'relative_temp', palette_paired[1], 'Climate Change', 'Temp. C', 'Date')

# line variables
y_0 = ax_1.get_ylim()[0]
y_1 = ax_1.get_ylim()[1]

# vline & hline
ax_1.vlines(x=pd.Timestamp('2015-10-06'), ymin=y_0, ymax=y_1, linewidth=1, linestyle='--', color='grey')

# annotate
ax_1.text(pd.Timestamp('2005-10-06'), 1.25, '2015-10-06: Temp > 1', color='grey')

# show
plt.show()
```

```python
# fig
fig, ax_1 = plt.subplots(nrows=1, ncols=1, figsize=(12,6), constrained_layout=True)

# viz
viz_line(ax_1, cc, 'relative_temp', palette_paired[1], 'Climate Change', 'Temp. C', 'Date')

# line variables
x_0 = ax_1.get_xlim()[0]
x_1 = ax_1.get_xlim()[1]
t_mean = cc['relative_temp'].mean()

# hline
ax_1.hlines(y=t_mean, xmin=x_0, xmax=x_1, linewidth=1, linestyle='--', color='grey')

# annotate
ax_1.text(pd.Timestamp('1956-01-01'), 0.35, 'Mean temp.: 0.3', color='grey')

# show
plt.show()
```
